# 편미분

- 식 안에 여러 변수가 존재
    - x1, x2, x3,.. ..y, ...
- 원하는 한가지 변수만 미분한다면
- **나머지 모든 변수는 상수로 취급**한다

- 식
    - f(x,y) = x^2 + yx + a(상수)
- 변수
    - x, y
- 위의 식 기준 변수 x에 대해서 미분하시오 => 편미분
    - 함수 f를 x에 관해 편미분하시오
    - δf / δx
- 편미분 결과
    - δf / δx = 2x^(2-1) + y*1 + 0
    - δf / δx = 2x + y

## 최소제곱법

- 잘 그은 예측선을 찾으면, 미래를 잘 예측할 수 있다

```python
import numpy as np
import pandas as pd

# 더미데이터
# hour : 운동시간
# score: 체력등급점수
df = pd.DataFrame({
    'hour'  : [3, 5, 7, 9],
    'score' : [82, 94, 92, 98]
})
df
```

```python
# 산포도
import matplotlib.pyplot as plt

plt.scatter( df.hour, df.score );
```

- 왼쪽아래에서 오른족 위로 향하는 선형 분포가 대략적 보임
- 일차함수 그래프 예상
- y = ax + b
    - x:독립변수, y:종속변수
    - 정확한 예측을 위해서는 a(기울기), b(y절편)을 구해야한다(최적으로)
        - 회귀계수
        - 선형회귀는 최적의 a,b를 구함

- 회귀계수 추정방식
    - OLS
        - 최소제곱법(최소자승법)

    - MLE
        - 최대우도법(최대가능도방법)

```python
# 기울기 a
Image('/content/drive/MyDrive/SeSAC/share/딥러닝/img/기초_최소제곱법을이용한기울기.jpg')
# 분자에 mean(x)) <- 가로 하나 누락
```

- 분모 : x의 편차(각각 x값과 평균x 간의 차이)를 제곱해서 합한값
- 분자 : x의 편차*y의 편차해서 모두 합한값4

```python
# 평균
mean_x = np.mean( df.hour )
mean_y = np.mean( df.score )

mean_x, mean_y
```

```python
# 최소 제곱법에 맞춰서 기울기 계산
# 분모
x       = df.hour.values
div_sub = sum( [ (i- mean_x)**2  for i in x ] )
div_sub

# 분자
y       = df.score.values
div_top = sum( [ (x[i] - mean_x)*(y[i] - mean_y) for i in range( len(y) ) ] )
div_top

# 기울기 a
a = div_top / div_sub
a
```

```python
# y 절편
#Image('/content/drive/MyDrive/SeSAC/share/딥러닝/img/기초_최소제곱법을이용한절편.jpg')
# b = y평균 - (x평균*기울기)
```

```python
b = mean_y - (mean_x * a)
b
```

```python
# 산포드 + 회귀선
# 회귀식
reg = lambda x:a*x + b

plt.scatter( df.hour, df.score )
plt.plot( df.hour, reg(df.hour) )

plt.ylim( 80, 118) # seaborn 형태와 유사하게 조정
plt.show()
```

## 평균 제곱 오차 (MSE)

- 회귀 알고리즘의 평가 지표:손실함수중 하나
- 피처가 많아지면, 최소제곱법으로 해결 불가
- 아이디어
    - 적당한 값(학습계수) 부여 -> 조금씩 수정
    - 적당한 값에 파생결과로 오차값이 최소가 되는지 체크 -> 수정 -> 반복...
    - 오차가 최소로 되는 지점을 찾을때까지 반복
    - 이때 판단 기준은 MSE,... 같은 지표 사용

```python
# 더미데이터를 위한 초기 값 부여 => 딥러닝에서는 초기 가중치값(w, b)
dum_a = 3   # 기울기
dum_b = 75  # y절편
# 회귀식
predict = lambda x:dum_a * x + dum_b
# 예측값
y_pred = [ predict( x[i] ) for i in range( len(x) )]
y_pred
```

```python
df['predict'] = y_pred
df
```

```python
# 오차값 계산 error 컬럼 추가
# score - predict
df['error'] = df.score.values - y_pred
df

# 에러값이 양수 혹은 음수로도 존재 => 오차량 사용 (절대값 혹은 제곱 혹은 제곱후 나중에 제곱근처리,..)
```

```python
Image('/content/drive/MyDrive/SeSAC/share/딥러닝/img/기초_MSE.jpg')
```

```python
# 평균 제곱 오차를 구현
n = len(x)
def mse(y, y_pred):
    return (1/n) * sum( (y - y_pred)**2 )

mse( y, y_pred ), np.sqrt( mse( y, y_pred ) ) # RMSE
```

```python
# 위의값 기반 산포도 + 회귀선
plt.scatter( df.hour, df.score )
plt.plot( df.hour, df.predict )

plt.ylim( 80, 118) # seaborn 형태와 유사하게 조정
plt.show()

# 회귀선이 얼추 그려졌다 => 최적의 회귀선이라 할수 없다 => a,b를 조정
```

```python
# 기울기와 오차값의 관계
# 기울기가 크거나, 작아지면 오차값을 커진다

a_set = [ -100, -10, -3, 0, 3, 10, 100] # 기울기를 모아놓은 것
plt.scatter( df.hour, df.score )        # 데이터 분포도
# 기울기를 바꿔가면서 회귀선 드로잉
for a in a_set:
    predict = lambda x: a*x + dum_b    # 기울기만 체크(y절편 체크 생략)
    y_pred  = predict(df.hour)
    plt.plot( df.hour, y_pred, label=f'{a}-line' ) # 회귀선

plt.legend()
plt.show()

# 기울기아 아주 크거나 작으면 오차값 상승, 적절 기울기면 오차값이 줄어든다
# 기울기와 오차는 2차함수 관계, 포물선(2차함수) 그래프 형태를 띤다
```

```python
n = len(x)              # 데이터개수
# 기본 학습 형태
for i in range(epochs):  # 3000번 학습 -> 최적의 a,b를(w,b)찾는 과정
    y_pred = a * x + b   # 단순선형회귀식
    error  = y - y_pred  # 오차값 (실제값-예측값)

    # 업데이트할 값 a, b 각각 준비 => a 혹은 b의 변동량 -> 편미분값
    a_update = (2/n) * sum(-x*(error))
    b_update = (2/n) * sum(-(error))

    # 업데이트 진행
    a = a - lr * a_update # a의 변동량의 학습률만큼만 적용, 업데이트 수행
    b = b - lr * b_update # b의 변동량의 학습률만큼만 적용, 업데이트 수행

    # 로그 출력(verbose)
    if i % 100 == 0:
        print( f'epoch={i:4} \t 기울기={a:0.4f} \t 절편={b:0.4f}')
    pass
```

```python
# 기울기, 절편
a1 = 0
a2 = 0
b  = 0
# 학습률
lr = 0.01
# 반복
epochs = 3000
```

```python
n = len(x1)              # 데이터개수

for i in range(epochs):
    y_pred = a1 * x1 + a2 * x2 + b   # 다중선형회귀식
    error  = y - y_pred              # 오차값 (실제값-예측값)

    a1_update = (2/n) * sum(-x1*(error))
    a2_update = (2/n) * sum(-x2*(error))
    b_update  = (2/n) * sum(-(error))

    a1 = a1 - lr * a1_update # a1의 변동량의 학습률만큼만 적용, 업데이트 수행
    a2 = a2 - lr * a2_update # a2의 변동량의 학습률만큼만 적용, 업데이트 수행
    b  = b - lr * b_update   # b의 변동량의 학습률만큼만 적용, 업데이트 수행

    # 로그 출력(verbose)
    if i % 100 == 0:
        print( f'epoch={i:4} \t 기울기={a1:0.4f} 기울기={a2:0.4f} \t 절편={b:0.4f}')
    pass
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
# 단일 퍼셉트론 구현 : 입력층 1개, 출력층 1개
model.add(Dense(1, input_dim=1, activation='linear'))
# 컴파일 옵션, 최적화옵션
model.compile( optimizer='sgd', loss='mse')
# 학습
model.fit( df2.hour, df2.score, epochs=2000)
```
