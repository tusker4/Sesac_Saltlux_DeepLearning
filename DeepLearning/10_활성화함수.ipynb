# Activation Function

- 사용 이유
    - 인공신경망의 성능 향상 목적
        - 최적화 도구
            - GD, SGD, Adam, Momentom,....
        - 활성화 함수
            - 층을 통과한 데이터를 비선형으로 조정
                - 데이터가 비선형으로 분포되 있으면, 역전파 과정에 미분값을 통해서 학습을 수행할수 있다 -> 최적화 수행 가능
                - 선형구조면 -> 미분시 -> 상수가 발생 => 학습 X
            - 정규화 처리
                - 노이즈 제거
                - 필요시 음수 제거
            - 특수 목적
                - 이진/다중 분류

- 단점
    - 정보 손실
    - 죽은 뉴런 문제
        - 가중치 업데이트 X
        - Vanishing Gradient 문제(기울기 소멸)
- 장점
    - 비선형구조 -> 신경망을 더 깊게 구성 가능하다
    - 이상치, 결측치 제거 역활도 담당(경우에따라)
    - 데이터 분포 조정
    - 출력층에 최종 보정

# sigmoid

- 로지스틱회귀때 부터 유래, 사용
- 입력
  - 실수값( - 무한대 ~ + 무한대)
- 출력
  - 0 or 1
- 특징
  - 이진분류
- 문제점
  - x값이 커지거나, 작아지면 0과 1에 수렴
    - x축과 평행 => 기울기 0 => 학습 X => 가중치 업데이트 X => 죽은 뉴런 => 기울기 소멸문제 발생
