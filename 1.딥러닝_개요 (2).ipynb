## 모델구축-(인공신경망)

- 데이터와 목적 -> 기본형 인공신경망
    - 영상(비전), 이미지 => 비전 => CNN, ...
    - 자연어, 시계열(금융,스마트팩토리) => 시퀀스(순서) => RNN,.., transformer(인코더 + 디코더) 계열
    - 생성 => GAN, .., 스테블디퓨전/gpt 계열
        - 인코더 + 디코더
    - ...
- 인공신경망 - 퍼셉트론 기반
    - 입력층    
    - 출력층
- **인공신경망 - 다중퍼셉트론 기반**
    - 입력층
    - 은닉층(중간층) : 층이 깊어지니까 성능이 좋아지더라. => 깊게 => 딥
    - 출력층

```python
# 케라스 스타일 1번 유형 Sequential에 추가(add)하는 스타일
from tensorflow.keras.models import Sequential

# 추가되는 것 => XXX 층(layer), 기본 유닛 : Dense
from tensorflow.keras.layers import Dense
```

- 인공신경망 설계 도면(?) 필요
    - 입력층 + 은닉층(중간층) + 출력층
        - shape 설계
        - 여기서는 임의 설정
    - 입력층과 첫번째 은닉층은 붙여서 표현(스타일)
    - 데이터 :
        - train_x (352, 17)
        - test_x  (118, 17)
        - y : (352,), (118,)
        - shape을 보고 피처수를 잘 체크함!!

```python
#
model = Sequential()

# input_dim: 입력 피처수:17 => 몇개로 확산(임의표현, 설정) => 30(설정) : 입력층으로 이해
# 30 : 출력수 : 은닉층으로 이해, 30은 설정
# 17개의 피처가 들어와서 30개로 확장 전파 => 17*30 => 510개의 w가 필요하다
# 활성화 함수 사용하여 데이터좀 조정 => 비선형 처리(망을 깊게 만들기위해서), 특징 강화, 정규화
model.add( Dense( 30, input_dim=17, activation='relu' ) )
# add() => 레이어를 결합 => 인공신경망에 연결했다

# 출력층 1개 생성해서 인공신경망에 연결
# 이 데이터는 생존 1 , 사망 0 => 이진분류 => 로지스틱회귀 => sigmoid  활성화 함수
# 1 : 출력을 1개로 수렴
model.add( Dense( 1, activation='sigmoid') )
'''
    - 이 인경망을 통해서 데이터를 학습
    - 훈련 시킬 최종 파라미터수는 571 개
        - 17*30 = 510(가중치:w) + 편향(30개) = 540
        - 1(가중치 w) + 편향(30개 = 31
        - 510 + 31 = 571
    - 참고 : GPT-3의 파라미터수 1750억개
    -        GPT-3.5 ?개 => 120억 소요
    -        GPT-4   1조개
'''
model.summary() # 인공신경망 요약
```

```python
model = Sequential()
model.add( Dense( 30, input_dim=17, activation='relu' ) ) # 30*17(w) + 30(b)
model.add( Dense( 1, activation='sigmoid') )              # 1(w) + 30(b) => CNN에서 좀더 구체적
# 인공신경망 레이어 구조 + 학습파라미터수(최적화를통해 최적의 예측선을 찾는 w) 요약
model.summary()
```

- 해석
    - 1f layer
        - 입력층
        - x1, x2 데이터를 NAND, OR 연산을 하는 층으로 각각 전송
    - 2f layer
        - 은닉층
        - NAND : 하나라도 0이 있으면 1
            - 입력에서 은닉을 데이터가 전달될때 가중치값이 곱해져서 결과가 NAND와 동일하게됨
        - OR   : 하나라도 1이 있으면 1
            - 입력에서 은닉을 데이터가 전달될때 가중치값이 곱해져서 결과가 OR와 동일하게됨
    - 3f layer
        - 출력층
        - NAND, OR의 입력을 AND
        - AND : 둘다 1이여야 1
        - 은닉에서 출력으로 데이터가 전달될때 가중치값이 곱해져서 결과가 AND와 동일하게됨
