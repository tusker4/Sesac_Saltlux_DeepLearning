## RNN (Recurrent Netural Network)
- 은닉층에 입력이 2개
  - Xt   : 현재 시점의 입력
  - Ht-1 : 이전 시점(직전)의 은닉층의 상태가 입력
- 과거의 입력(은닉층의 상태)이 현재의 은닉층에 영향을 미친다
  - 과거의 데이터(정보)가 현재에 영향을 미친다
  - 기억(메모리) => 과거의 기억을 계속 유지한다
- 단점
  - 과거가 아주 오래되었다면, 초기에 영향을 미친 정보들이 희소해진다.
    - **장기기억소실문제** 발생
  - 이로인해, 오차역전파 수행시
    - 기울기 폭주 (Gradient Exploding)
    - 기울기 소실 (Gradient Vanishing)
- 종류
  - 기본 모형 : Vanilla RNN, SimpleRNN,
